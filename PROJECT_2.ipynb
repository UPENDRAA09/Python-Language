ii. Outliers

Generate a summary of count of all the outliers column wise
Replace all outlier values for numeric columns by mean. 
(Hint1: for outlier treatment use IQR method as follows:

For example: for a column X calculate Q1 = 25th percentile and Q3 = 75th percentile then IQR = Q3 â€“ Q1 ) then to check outlier, anything lower than a Q1 â€“ 1.5IQR or greater than Q3 + 1.5 IQR would be an outlier

Hint2: For getting percentile value, explore pd.describe() function)

```python
DataFrame1.describe()
```

```python
DataFrame2.describe()
```

```python
def findout_outliers(data_frame,col_name):
    global Q3,Q1,IQR_median,lower_bound,upper_bound
    Q1=data_frame[col_name].quantile(.25)
    Q3=data_frame[col_name].quantile(.75)
    IQR_median=(Q3-Q1)
    lower_bound=Q1-1.5*IQR_median
    upper_bound=Q3+1.5*IQR_median
    return ((data_frame[col_name]<lower_bound) | (data_frame[col_name]>upper_bound)).sum()
print("="*50)
print("Customer Details Dataset DataFrame:")
print("Customer_details_df with age:",findout_outliers(DataFrame1,'age'))
print("Customer_details_df with driving licence present:",findout_outliers(DataFrame1,'driving licence present2'))
print("Customer_details_df with region code:",findout_outliers(DataFrame1,'region code'))
print("Customer_details_df with previously_insured:",findout_outliers(DataFrame1,'previously insured'))
print("="*50)
print("Customer Policy  Details Dataset DataFrame:")
print("customer_policy_details_df with annual_premium:",findout_outliers(DataFrame2,'annual premium (in Rs)'))
print("customer_policy_details_df with sales channel code:",findout_outliers(DataFrame2,'sales channel code'))
print("customer_policy_details_df with vintage:",findout_outliers(DataFrame2,'vintage'))
print("customer_policy_details_df with response:",findout_outliers(DataFrame2,'response'))
print("="*50)
```

```python
#-Replace all outlier values for numeric columns by mean.
def replace_outliers(data_frame,col_name):
    findout_outliers(data_frame,col_name)
    outliers= ~data_frame[col_name].between(lower_bound,upper_bound)
    Average=data_frame.loc[~outliers,col_name].mean()
    data_frame.loc[outliers,col_name]=Average
    return ((data_frame[col_name]<lower_bound) | (data_frame[col_name]>upper_bound)).sum()
print("="*50)
print("Customer Details Dataset DataFrame:")
print("Customer_details_df with age:",replace_outliers(DataFrame1,'age'))
print("Customer_details_df with driving licence present:",replace_outliers(DataFrame1,'driving licence present2'))
print("Customer_details_df with region code:",replace_outliers(DataFrame1,'region code'))
print("Customer_details_df with previously_insured:",replace_outliers(DataFrame1,'previously insured'))
print("="*50)
print("Customer Policy  Details Dataset DataFrame:")
print("customer_policy_details_df with annual_premium:",replace_outliers(DataFrame2,'annual premium (in Rs)'))
print("customer_policy_details_df with sales channel code:",replace_outliers(DataFrame2,'sales channel code'))
print("customer_policy_details_df with vintage:",replace_outliers(DataFrame2,'vintage'))
print("customer_policy_details_df with response:",replace_outliers(DataFrame2,'response'))
print("="*50)

```

Hint: use correlation function (Correlation describes the relationship between two variables). 

Correlation coefficient < -0.5           - Strong negative relationship

Correlation coefficient > 0.5            -  Strong positive relationship

0.5 < Correlation coefficient < 0.5   - There is no relationship. 

```python
coo1 = Master_Table['age'].corr(Master_Table['annual premium (in Rs)'])
#print('there is no relationship=>',coo1)
if coo1 < -0.5 :
    print('Strong negative relationship')
elif coo1 > 0.5:
    print('Strong positive relationship')
elif coo1>-0.5 and coo1<0.5:
    print('There is no relationship')
else:
    print('check your value')
```
